```python
data1 = {'Date': ['2023-01-01', '2023-01-01', '2023-01-02', '2023-01-02'],
 'City': ['New York', 'Los Angeles', 'New York', 'Los Angeles'],
 'Temperature': [32, 75, 30, 77],
 'Humidity': [80, 10, 85, 5]}

df2= pd.DataFrame(data1)

pivot_df = df2.pivot_table(index=['City','Temperature'], columns='Date', 
values='Humidity',aggfunc="mean")
print(pivot_df)
```
